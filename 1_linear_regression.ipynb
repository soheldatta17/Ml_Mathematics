<h3 style="color:purple">Sample problem of predicting home price in monroe, new jersey (USA)</h3>

Dataset link: https://github.com/soheldatta17/Ml_Projects/blob/main/Day_1/canada_per_capita_income.csv

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/canada_per_capita_income.csv')
df
```

```python
df.rename(columns={'year': 'area', 'per capita income (US$)': 'price'}, inplace=True)

plt.figure(figsize=(10, 6))
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Price vs Area')
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.grid(True)
plt.show()
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

**(1) Predict price of a home with area = 3300 sqr ft**

**(2) Predict price of a home with area = 2005 sqr ft**

```python
reg.predict([[3300]])
```

```python
reg.predict([[3300]])
```

```python
reg.predict([[2005]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
